```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv("Telco-Customer-Churn.csv")

df.head()
```

```python
# criando um dicionário de tradução para as colunas numéricas
colunas_para_traduzir = {
    'SeniorCitizen': 'Idoso',
    'tenure': 'Meses_na_Empresa',
    'MonthlyCharges': 'Valor_Mensal',
    'TotalCharges': 'Valor_Total'
}

# renomeando no DataFrame
df.rename(columns=colunas_para_traduzir, inplace=True)
```

```python
df['Valor_Total'] = pd.to_numeric(df['Valor_Total'], errors='coerce')
```

```python
print(f"Valores nulos encontrados: {df['Valor_Total'].isnull().sum()}")
```

```python
df['Valor_Total'] = df['Valor_Total'].fillna(0)
```

```python
print(df['Valor_Total'].mean())
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Contract', hue='Churn')
plt.title('Cancelamento por Tipo de Contrato')
plt.xlabel('Tipo de Contrato')
plt.ylabel('Quantidade de Clientes')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='OnlineSecurity', hue='Churn')
plt.title('Impacto da Segurança Online no Cancelamento')
plt.show()
```

```python
# distribuição do tempo de contrato (Meses) por cancelamento
plt.figure(figsize=(10,6))
sns.kdeplot(df[df['Churn'] == 'Yes']['Meses_na_Empresa'], label='Cancelou', fill=True, color="red")
sns.kdeplot(df[df['Churn'] == 'No']['Meses_na_Empresa'], label='Ficou', fill=True, color="blue")
plt.title('Distribuição do Tempo de Contrato (Meses) por cancelamento')
plt.xlabel('Meses na Empresa (Permanência)')
plt.legend()
plt.show()
```

```python
# transformar Churn em número para aparecer no mapa (Yes=1, No=0)
df['Cancelou_Num'] = df['Churn'].apply(lambda x: 1 if x == 'Yes' else 0)

colunas_mapa = ['Idoso', 'Meses_na_Empresa', 'Valor_Mensal', 'Valor_Total', 'Cancelou_Num']
matriz_correlacao = df[colunas_mapa].corr()


plt.figure(figsize=(10, 8))
sns.heatmap(matriz_correlacao, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Mapa de Correlação: Factores de Cancelamento')
plt.show()
```

```python
# serviços a serem analisados:
servicos = ['OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport']

# estrutura para armazenar os resultados:
resultados = []

for servico in servicos:
    # calcular a taxa de churn para quem NÃO tem o serviço (No)
    taxa = df[df[servico] == 'No']['Churn'].value_counts(normalize=True)['Yes'] * 100
    resultados.append({'Serviço': servico, 'Taxa_de_Cancelamento': taxa})

# DataFrama para facilitar o gráfico:
df_ranking = pd.DataFrame(resultados).sort_values(by='Taxa_de_Cancelamento', ascending=False)

# gráfico de barras:
plt.figure(figsize=(10, 6))
sns.barplot(data=df_ranking, x='Taxa_de_Cancelamento', y='Serviço', palette='Reds_r')

plt.title('Taxa de Cancelamento entre Clientes SEM o Serviço')
plt.xlabel('Porcentagem de Churn (%)')
plt.ylabel('Serviço Faltante')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```
